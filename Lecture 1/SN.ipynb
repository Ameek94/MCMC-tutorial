```python
import matplotlib
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import sys, os
sys.path.insert(0,os.path.realpath(os.path.join(os.getcwd(),'..')))
from getdist import plots, MCSamples
import getdist
import matplotlib.pyplot as plt
import IPython
print('GetDist Version: %s, Matplotlib version: %s'%(getdist.__version__, plt.matplotlib.__version__))
from getdist import loadMCSamples
matplotlib.rc('font', size=18)
matplotlib.rc('legend', fontsize=16)
matplotlib.rc('text', usetex=True) # remove if latex not installed 
matplotlib.rc('text.latex', preamble=r'\usepackage{lmodern}') # remove if latex not installed
```

```python
samples = loadMCSamples(os.path.join(os.getcwd(),'./chains/Sn')
                           ,settings={'ignore_rows':0.3})

keys = ['omegam','omega_de','w','H0']
for key in keys:
    print(samples.getInlineLatex(key,limit=2))
print(f" R-1 = {samples.getGelmanRubin():.4f} ")

```

#### Now we plot the contours of the 2d distributions. These are the regions that enclose 68% and 95% (or your choice of fraction) of the total probability density

```python
g = plots.get_subplot_plotter(subplot_size=2.)
g.settings.axes_fontsize=16
g.settings.legend_fontsize = 16
g.settings.axes_labelsize = 16
g.settings.title_limit_fontsize = 12
keys = ['omegam','omega_de','w']
g.triangle_plot(samples,keys
                ,filled=True,title_limit=2)
```
